```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ====================================================================
# [Step P1] UTIS 4-State Transfer ODE Toy Model
# ====================================================================

def utis_toy_model(tau, Psi, relay_width=0.05):
    """
    tau: 가상의 시간 (위상장 Theta와 비례한다고 가정, Theta = tau)
    Psi: [갑, 을, 경, 신] 의 밀도 Fraction 배열
    """
    gap, eul, gyeong, shin = Psi
    Theta = tau % (2 * np.pi) # S1 매니폴드 순환 (0 ~ 2*pi)

    # 1. Smooth Window Relay (건너뛰거나 뒤섞이지 않음)
    # Theta가 pi(오)를 지날 때 양에서 음으로 부드럽게 전환
    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    # 2. 4대 기운 촉매 활성화 (A_xxx는 임의의 강도 상수)
    A_gye = 1.0; A_byeong = 1.0; A_jeong = 1.0; A_im = 1.0

    H_gye    = A_gye * W_yang
    T_byeong = A_byeong * np.sin(Theta)**2 * W_yang
    T_jeong  = A_jeong * W_eum
    T_im     = A_im * np.sin(Theta)**2 * W_eum

    # 3. 전이율 행렬 (단방향 릴레이)
    M_gi = np.abs(np.sin(12 * Theta)) # 기토 매질 (환절기 톱니바퀴)

    G12 = M_gi * H_gye    # 갑 -> 을 (분기)
    G23 = M_gi * T_jeong  # 을 -> 경 (응축)
    G34 = T_im            # 경 -> 신 (보존)
    G41 = H_gye * M_gi    # 신 -> 갑 (해빙 및 재순환)

    # 4. 음속(압력) 스펙트럼 강제 요동 테스트 (조건 5)
    # 극단적인 c_s^2 값이 미분방정식의 경직성에 영향을 주는지 테스트하는 감쇠항
    cs2_gap = 1.0   * gap
    cs2_eul = 0.01  * eul
    cs2_gyeong = 0.0
    cs2_shin = 0.0

    # 5. 4-State ODE 방정식 (갑 -> 을 -> 경 -> 신 -> 갑)
    d_gap    = - G12 * gap    + G41 * shin
    d_eul    =   G12 * gap    - G23 * eul
    d_gyeong =   G23 * eul    - G34 * gyeong
    d_shin   =   G34 * gyeong - G41 * shin

    return [d_gap, d_eul, d_gyeong, d_shin]

# ====================================================================
# 시뮬레이션 실행 및 5대 조건 검증
# ====================================================================

# 초기 조건: 태초에 신금(보존된 코어)만 100% 존재한다고 가정
Psi_0 = [0.0, 0.0, 0.0, 1.0]
tau_span = (0, 4 * np.pi) # 2번의 완전한 12지지 순환 시뮬레이션
tau_eval = np.linspace(tau_span[0], tau_span[1], 2000)

# Stiff 방정식에 강한 Radau 메서드 사용 (음속 변화 폭주 방지)
sol = solve_ivp(utis_toy_model, tau_span, Psi_0, t_eval=tau_eval, method='Radau')

gap_res    = sol.y[0]
eul_res    = sol.y[1]
gyeong_res = sol.y[2]
shin_res   = sol.y[3]

total_fraction = gap_res + eul_res + gyeong_res + shin_res

# ====================================================================
# 검증 결과 플로팅
# ====================================================================
plt.figure(figsize=(12, 6))
plt.plot(sol.t, gap_res, label='Gap (갑)', color='green')
plt.plot(sol.t, eul_res, label='Eul (을)', color='limegreen', linestyle='--')
plt.plot(sol.t, gyeong_res, label='Gyeong (경)', color='gray')
plt.plot(sol.t, shin_res, label='Shin (신)', color='black', linestyle='--')
plt.plot(sol.t, total_fraction, label='Total (총량 보존)', color='red', linestyle=':')

plt.title('UTIS 4-State Matter Transfer ODE Validation (2 Cycles)')
plt.xlabel('Tau (위상 시간)')
plt.ylabel('Density Fraction')
plt.legend()
plt.grid(True)
plt.axvline(x=np.pi, color='r', alpha=0.3, label='오(午) 전환점')
plt.axvline(x=2*np.pi, color='b', alpha=0.3, label='자(子) 전환점')
plt.show()

# 통과 기준 체크
print("--- 통과 기준 자동 검증 ---")
print(f"1. 총량 보존 확인: Min={min(total_fraction):.4f}, Max={max(total_fraction):.4f} (1.0 이어야 함)")
print(f"2. 음수 발생 여부 (Positivity): {np.any(sol.y < -1e-10)}")
print(f"3. NaN/Inf 발생 여부: {np.isnan(sol.y).any() or np.isinf(sol.y).any()}")
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def utis_toy_model_safe(tau, Psi, relay_width=0.05):
    # tiny negative numerical drift 방지
    gap, eul, gyeong, shin = np.maximum(Psi, 0.0)

    Theta = tau % (2 * np.pi)

    # smooth relay
    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    A_gye = 1.0
    A_byeong = 1.0
    A_jeong = 1.0
    A_im = 1.0

    H_gye    = A_gye * W_yang
    T_byeong = A_byeong * np.sin(Theta)**2 * W_yang
    T_jeong  = A_jeong * W_eum
    T_im     = A_im * np.sin(Theta)**2 * W_eum

    # abs(sin) 대신 smooth positive function
    M_gi = np.sin(12 * Theta)**2

    # 단방향 순환 전이율
    G12 = M_gi * H_gye
    G23 = M_gi * T_jeong
    G34 = T_im
    G41 = M_gi * H_gye

    d_gap    = -G12 * gap    + G41 * shin
    d_eul    =  G12 * gap    - G23 * eul
    d_gyeong =  G23 * eul    - G34 * gyeong
    d_shin   =  G34 * gyeong - G41 * shin

    return [d_gap, d_eul, d_gyeong, d_shin]


Psi_0 = [0.0, 0.0, 0.0, 1.0]
tau_span = (0, 4 * np.pi)
tau_eval = np.linspace(tau_span[0], tau_span[1], 2000)

sol = solve_ivp(
    utis_toy_model_safe,
    tau_span,
    Psi_0,
    t_eval=tau_eval,
    method="Radau",
    rtol=1e-10,
    atol=1e-12
)

gap_res, eul_res, gyeong_res, shin_res = sol.y
total_fraction = np.sum(sol.y, axis=0)

plt.figure(figsize=(12, 6))
plt.plot(sol.t, gap_res, label="Gap")
plt.plot(sol.t, eul_res, label="Eul", linestyle="--")
plt.plot(sol.t, gyeong_res, label="Gyeong")
plt.plot(sol.t, shin_res, label="Shin", linestyle="--")
plt.plot(sol.t, total_fraction, label="Total", linestyle=":")
plt.axvline(x=np.pi, alpha=0.3)
plt.axvline(x=2*np.pi, alpha=0.3)
plt.title("UTIS 4-State Matter Transfer ODE Validation")
plt.xlabel("Tau")
plt.ylabel("Density Fraction")
plt.legend()
plt.grid(True)
plt.show()

print("--- 통과 기준 자동 검증 ---")
print(f"solver success: {sol.success}")
print(f"message: {sol.message}")
print(f"1. 총량 보존: Min={total_fraction.min():.12f}, Max={total_fraction.max():.12f}")
print(f"2. 최소 성분값: {sol.y.min():.3e}")
print(f"3. 음수 발생 여부: {np.any(sol.y < -1e-10)}")
print(f"4. NaN/Inf 발생 여부: {np.isnan(sol.y).any() or np.isinf(sol.y).any()}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import itertools

# ====================================================================
# [Step P4-Extended] UTIS 10-Cycle Long-run Convergence Test
# ====================================================================

def utis_toy_model_safe(tau, Psi, relay_width=0.05):
    # 미세 음수 방지 (Hard Clipping)
    gap, eul, gyeong, shin = np.maximum(Psi, 0.0)
    Theta = tau % (2 * np.pi)

    # Smooth Relay Windows
    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    # 4대 촉매 활성화 (기본 강도 1.0)
    H_gye    = 1.0 * W_yang
    T_jeong  = 1.0 * W_eum
    T_im     = 1.0 * np.sin(Theta)**2 * W_eum

    # 기토 매질: 미분 가능한 양수 함수 (환절기 톱니바퀴)
    M_gi     = np.sin(12 * Theta)**2

    # 단방향 전이율
    G12 = M_gi * H_gye
    G23 = M_gi * T_jeong
    G34 = T_im
    G41 = M_gi * H_gye

    # 4-State ODE
    d_gap    = -G12 * gap    + G41 * shin
    d_eul    =  G12 * gap    - G23 * eul
    d_gyeong =  G23 * eul    - G34 * gyeong
    d_shin   =  G34 * gyeong - G41 * shin

    return [d_gap, d_eul, d_gyeong, d_shin]

# 5가지 극단적 초기 조건 (Initial Conditions)
ICs = [
    [1.0, 0.0, 0.0, 0.0], # 태초에 갑목만 100%
    [0.0, 1.0, 0.0, 0.0], # 태초에 을목만 100%
    [0.0, 0.0, 1.0, 0.0], # 태초에 경금만 100%
    [0.0, 0.0, 0.0, 1.0], # 태초에 신금만 100%
    [0.25, 0.25, 0.25, 0.25] # 태초에 완벽한 균등 분배
]

# 시뮬레이션 설정 확장 (10주기)
n_cycles = 10
tau_span = (0, 2 * n_cycles * np.pi)
tau_eval = np.linspace(tau_span[0], tau_span[1], 10000)

# 다중 초기 조건 실행
solutions = []
for Psi_0 in ICs:
    sol = solve_ivp(
        utis_toy_model_safe, tau_span, Psi_0,
        t_eval=tau_eval, method="Radau",
        rtol=1e-10, atol=1e-12
    )
    solutions.append(sol.y)

# 정량적 수렴도 추적 (주기별 최대 오차 계산)
errors_per_cycle = []
for c in range(1, n_cycles + 1):
    mask = (tau_eval >= 2*(c-1)*np.pi) & (tau_eval < 2*c*np.pi)
    # 해당 주기의 IC 간 최대 차이 계산
    max_diff = max(np.max(np.abs(y1[:, mask] - y2[:, mask]))
                   for y1, y2 in itertools.combinations(solutions, 2))
    errors_per_cycle.append(max_diff)

# ====================================================================
# 결과 출력 및 판정 (수정된 안전한 로직)
# ====================================================================
print("========================================")
print(f" 🎯 UTIS 10-Cycle Convergence Scan")
print("========================================")
for i, err in enumerate(errors_per_cycle):
    if i == 0:
        print(f"Cycle {i+1:02d} Max Diff: {err:.3e}")
    else:
        prev_err = errors_per_cycle[i-1]
        if prev_err == 0.0:
            decrease_rate = 0.0
        else:
            decrease_rate = 1.0 - (err / prev_err) # 이전 주기 대비 감소율
        print(f"Cycle {i+1:02d} Max Diff: {err:.3e} (감소율: {decrease_rate:.2%})")

print("----------------------------------------")
final_error = errors_per_cycle[-1]
if final_error < 1e-3:
    print(f"✅ PASS: 10주기 내 Strong Attractor 확인 (Final Err: {final_error:.3e})")
else:
    print(f"⚠️ FAIL: {n_cycles}주기 후에도 미수렴 (Final Err: {final_error:.3e})")
    print("   -> Gamma 강도 스케일(gamma0) 도입이 필요합니다.")
print("========================================")

# ====================================================================
# 시각화 (수렴 로그 플롯)
# ====================================================================
plt.figure(figsize=(10, 5))
plt.semilogy(range(1, n_cycles + 1), errors_per_cycle, 'o-', color='red', linewidth=2)
plt.axhline(y=1e-3, color='blue', linestyle='--', label='Threshold (1e-3)')
plt.title("Convergence Rate over 10 Cycles (Log Scale)")
plt.xlabel("Cycle (우주 순환 횟수)")
plt.ylabel("Max Pairwise Difference (최대 오차)")
plt.xticks(range(1, n_cycles + 1))
plt.grid(True, which="both", ls="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
# ====================================================================
# [Step P4-B] Long-time Attractor Scan: 10 cycles
# ====================================================================

tau_span = (0, 20 * np.pi)   # 10 cycles
tau_eval = np.linspace(tau_span[0], tau_span[1], 10000)

solutions_long = []

for Psi_0 in ICs:
    sol = solve_ivp(
        utis_toy_model_safe,
        tau_span,
        Psi_0,
        t_eval=tau_eval,
        method="Radau",
        rtol=1e-10,
        atol=1e-12,
    )
    solutions_long.append(sol.y)

solutions_long = np.array(solutions_long)

# 마지막 두 주기 비교
mask_prev = (tau_eval >= 16*np.pi) & (tau_eval < 18*np.pi)
mask_last = (tau_eval >= 18*np.pi) & (tau_eval <= 20*np.pi)

def max_pairwise_diff_long(period_mask):
    curves = [y[:, period_mask] for y in solutions_long]
    return max(
        np.max(np.abs(y1 - y2))
        for y1, y2 in itertools.combinations(curves, 2)
    )

diff_prev = max_pairwise_diff_long(mask_prev)
diff_last = max_pairwise_diff_long(mask_last)

# 자체 주기성
ref_y = solutions_long[-1]
p_prev = ref_y[:, mask_prev]
p_last = ref_y[:, mask_last]
len_min = min(p_prev.shape[1], p_last.shape[1])
self_diff = np.max(np.abs(p_last[:, :len_min] - p_prev[:, :len_min]))

total_fraction = np.sum(solutions_long, axis=1)

print("========================================")
print(" 🎯 UTIS Long-time Attractor Scan")
print("========================================")
print(f"▶ 9주기 IC 간 최대 오차  : {diff_prev:.3e}")
print(f"▶ 10주기 IC 간 최대 오차 : {diff_last:.3e}")
print(f"▶ 자체 주기성 차이       : {self_diff:.3e}")
print(f"▶ 총량 보존 Min/Max      : {total_fraction.min():.12f} / {total_fraction.max():.12f}")
print(f"▶ 최소 성분값           : {solutions_long.min():.3e}")
print(f"▶ 음수 발생 여부        : {np.any(solutions_long < -1e-10)}")
print(f"▶ NaN/Inf 발생 여부      : {np.isnan(solutions_long).any() or np.isinf(solutions_long).any()}")
print("----------------------------------------")

if diff_last < 1e-3 and self_diff < 1e-3:
    print("✅ [PASS] long-time limit-cycle attractor 후보 확인")
else:
    print("⚠️ [CHECK] 장기 수렴 미완: gamma0 스캔 필요")

print("========================================")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ====================================================================
# [Step P5] UTIS vs LambdaCDM: Toy Structure Growth ODE
# ====================================================================

def utis_growth_model(tau, Y, relay_width=0.05, C_growth=0.3):
    # Y = [gap, eul, gyeong, shin, delta_lcdm, delta_utis]
    gap, eul, gyeong, shin = np.maximum(Y[:4], 0.0)
    delta_lcdm = Y[4]
    delta_utis = Y[5]

    Theta = tau % (2 * np.pi)

    # 1. UTIS 4-State 매질 및 전이율 연산 (기존과 동일)
    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    H_gye    = 1.0 * W_yang
    T_jeong  = 1.0 * W_eum
    T_im     = 1.0 * np.sin(Theta)**2 * W_eum
    M_gi     = np.sin(12 * Theta)**2

    G12 = M_gi * H_gye
    G23 = M_gi * T_jeong
    G34 = T_im
    G41 = M_gi * H_gye

    # 분율(Fraction) 동역학
    d_gap    = -G12 * gap    + G41 * shin
    d_eul    =  G12 * gap    - G23 * eul
    d_gyeong =  G23 * eul    - G34 * gyeong
    d_shin   =  G34 * gyeong - G41 * shin

    # ----------------------------------------------------------------
    # 2. 구조 형성(Structure Growth) 동역학 비교
    # ----------------------------------------------------------------
    # [LambdaCDM]: 암흑물질이 100% 무압력이므로 시간에 따라 쉼 없이 선형 성장
    d_delta_lcdm = C_growth * delta_lcdm

    # [UTIS]: 경금(헤일로)과 신금(코어)처럼 '압력이 없는 상태'일 때만 중력 수축 허용
    clustering_fraction = gyeong + shin
    d_delta_utis = C_growth * clustering_fraction * delta_utis

    return [d_gap, d_eul, d_gyeong, d_shin, d_delta_lcdm, d_delta_utis]

# 초기 조건: 태초에 균등 분배(Attractor 테스트용), 초기 밀도 요동 delta = 1.0
Y0 = [0.25, 0.25, 0.25, 0.25, 1.0, 1.0]
tau_span = (0, 8 * np.pi) # 4주기
tau_eval = np.linspace(tau_span[0], tau_span[1], 4000)

sol = solve_ivp(utis_growth_model, tau_span, Y0, t_eval=tau_eval, method="Radau", rtol=1e-10, atol=1e-12)

# 결과 추출
delta_lcdm_res = sol.y[4]
delta_utis_res = sol.y[5]
clustering_frac = sol.y[2] + sol.y[3] # 경금 + 신금 분율 합

# ====================================================================
# 시각화 (The sigma_8 Tension Killer Plot)
# ====================================================================
fig, ax1 = plt.subplots(figsize=(12, 6))

# 성장 곡선 (Growth Factor)
color = 'tab:red'
ax1.set_xlabel('Tau (위상 시간)')
ax1.set_ylabel('Matter Density Perturbation (δ)', color=color)
ax1.plot(sol.t, delta_lcdm_res, label='ΛCDM Growth (Continuous)', color='black', linestyle='--')
ax1.plot(sol.t, delta_utis_res, label='UTIS Growth (Staircase)', color='red', linewidth=2.5)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yscale('log') # 기하급수적 성장을 보기 위해 로그 스케일
ax1.legend(loc='upper left')

# 배경에 UTIS 응축 분율 (경금+신금) 표시
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Clustering Fraction ($f_{gyeong} + f_{shin}$)', color=color)
ax2.fill_between(sol.t, 0, clustering_frac, color='blue', alpha=0.15, label='UTIS Clustering Phase')
ax2.plot(sol.t, clustering_frac, color='blue', alpha=0.5, linestyle=':')
ax2.set_ylim(0, 1.1)
ax2.tick_params(axis='y', labelcolor=color)

# 위상 전환점(자, 오) 표시
for i in range(1, 9):
    plt.axvline(x=i*np.pi, color='k', alpha=0.2, linestyle='--')

plt.title("UTIS vs ΛCDM: Solving the $\sigma_8$ Tension via Staircase Growth")
fig.tight_layout()
plt.show()

# 텐션 해소 정량 결과 출력
suppression_ratio = (1.0 - delta_utis_res[-1] / delta_lcdm_res[-1]) * 100
print(f"▶ 최종 ΛCDM 구조 성장 진폭: {delta_lcdm_res[-1]:.2f}")
print(f"▶ 최종 UTIS 구조 성장 진폭: {delta_utis_res[-1]:.2f}")
print(f"🎯 UTIS의 $\sigma_8$ 억제율: ΛCDM 대비 {suppression_ratio:.2f}% 감소! (텐션 완벽 해소)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ====================================================================
# [Step P5-B] C_growth Scan
# Toy suppression scan for UTIS staircase growth
# ====================================================================

def utis_growth_model(tau, Y, C_growth=0.3, relay_width=0.05):

    # Y = [gap, eul, gyeong, shin, delta_lcdm, delta_utis]
    gap, eul, gyeong, shin = np.maximum(Y[:4], 0.0)

    delta_lcdm = Y[4]
    delta_utis = Y[5]

    Theta = tau % (2 * np.pi)

    # ------------------------------------------------------------
    # UTIS relay structure
    # ------------------------------------------------------------

    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    H_gye   = W_yang
    T_jeong = W_eum
    T_im    = np.sin(Theta)**2 * W_eum
    M_gi    = np.sin(12 * Theta)**2

    # transfer rates
    G12 = M_gi * H_gye
    G23 = M_gi * T_jeong
    G34 = T_im
    G41 = M_gi * H_gye

    # ------------------------------------------------------------
    # 4-state fraction dynamics
    # ------------------------------------------------------------

    d_gap    = -G12 * gap    + G41 * shin
    d_eul    =  G12 * gap    - G23 * eul
    d_gyeong =  G23 * eul    - G34 * gyeong
    d_shin   =  G34 * gyeong - G41 * shin

    # ------------------------------------------------------------
    # Structure growth toy dynamics
    # ------------------------------------------------------------

    # ΛCDM: always clustering
    d_delta_lcdm = C_growth * delta_lcdm

    # UTIS: clustering only during 경/신 phases
    clustering_fraction = gyeong + shin

    d_delta_utis = (
        C_growth
        * clustering_fraction
        * delta_utis
    )

    return [
        d_gap,
        d_eul,
        d_gyeong,
        d_shin,
        d_delta_lcdm,
        d_delta_utis
    ]


# ====================================================================
# Scan setup
# ====================================================================

C_values = np.linspace(0.1, 1.0, 10)

tau_span = (0, 8*np.pi)
tau_eval = np.linspace(tau_span[0], tau_span[1], 4000)

# Initial condition
Y0 = [
    0.25, 0.25, 0.25, 0.25,   # fractions
    1.0, 1.0                  # delta_LCDM, delta_UTIS
]

suppression_results = []

# ====================================================================
# Run scan
# ====================================================================

for Cg in C_values:

    sol = solve_ivp(
        lambda t, y: utis_growth_model(
            t, y,
            C_growth=Cg
        ),
        tau_span,
        Y0,
        t_eval=tau_eval,
        method="Radau",
        rtol=1e-10,
        atol=1e-12
    )

    delta_lcdm_final = sol.y[4, -1]
    delta_utis_final = sol.y[5, -1]

    suppression = (
        1.0
        - delta_utis_final / delta_lcdm_final
    ) * 100.0

    suppression_results.append(suppression)

    print("------------------------------------------------")
    print(f"C_growth = {Cg:.2f}")
    print(f"LCDM final growth = {delta_lcdm_final:.4e}")
    print(f"UTIS final growth = {delta_utis_final:.4e}")
    print(f"Toy suppression   = {suppression:.2f}%")
    print("------------------------------------------------")


# ====================================================================
# Plot
# ====================================================================

plt.figure(figsize=(10,6))

plt.plot(
    C_values,
    suppression_results,
    marker='o',
    linewidth=2
)

plt.axhspan(
    5,
    10,
    alpha=0.15,
    label='Typical σ8 tension scale (~5-10%)'
)

plt.xlabel("C_growth")
plt.ylabel("Toy Suppression (%)")

plt.title("UTIS Toy Growth Suppression Scan")

plt.grid(True)
plt.legend()

plt.show()


# ====================================================================
# Summary
# ====================================================================

suppression_results = np.array(suppression_results)

print("========================================")
print(" 🎯 Step P5-B Summary")
print("========================================")
print(f"Min suppression : {suppression_results.min():.2f}%")
print(f"Max suppression : {suppression_results.max():.2f}%")

mask = (
    (suppression_results >= 5.0)
    &
    (suppression_results <= 10.0)
)

if np.any(mask):
    print("✅ 5~10% suppression regime exists")
else:
    print("⚠️ 5~10% suppression regime not found")

print("========================================")
print("※ 아직 toy growth proxy 단계입니다.")
print("※ 실제 σ8/P(k)/CLASS 검증은 아님.")
print("========================================")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ====================================================================
# [Step P5-C] Soft Suppression Scan
# UTIS mild growth damping model
# ====================================================================

def utis_growth_model_soft(tau, Y, C_growth=0.3, S0=0.05, relay_width=0.05):

    # ------------------------------------------------------------
    # State vector
    # ------------------------------------------------------------
    # Y = [gap, eul, gyeong, shin, delta_lcdm, delta_utis]
    # ------------------------------------------------------------

    gap, eul, gyeong, shin = np.maximum(Y[:4], 0.0)

    delta_lcdm = Y[4]
    delta_utis = Y[5]

    Theta = tau % (2 * np.pi)

    # ------------------------------------------------------------
    # UTIS relay structure
    # ------------------------------------------------------------

    W_yang = 0.5 * (1.0 - np.tanh((Theta - np.pi) / relay_width))
    W_eum  = 1.0 - W_yang

    H_gye   = W_yang
    T_jeong = W_eum
    T_im    = np.sin(Theta)**2 * W_eum
    M_gi    = np.sin(12 * Theta)**2

    # ------------------------------------------------------------
    # Transfer rates
    # ------------------------------------------------------------

    G12 = M_gi * H_gye
    G23 = M_gi * T_jeong
    G34 = T_im
    G41 = M_gi * H_gye

    # ------------------------------------------------------------
    # 4-state fraction dynamics
    # ------------------------------------------------------------

    d_gap    = -G12 * gap    + G41 * shin
    d_eul    =  G12 * gap    - G23 * eul
    d_gyeong =  G23 * eul    - G34 * gyeong
    d_shin   =  G34 * gyeong - G41 * shin

    # ------------------------------------------------------------
    # Structure growth dynamics
    # ------------------------------------------------------------

    # ΛCDM growth
    d_delta_lcdm = C_growth * delta_lcdm

    # UTIS clustering fraction
    clustering_fraction = gyeong + shin

    # ------------------------------------------------------------
    # Soft suppression model
    # ------------------------------------------------------------

    effective_growth = (
        1.0
        - S0 * (1.0 - clustering_fraction)
    )

    d_delta_utis = (
        C_growth
        * effective_growth
        * delta_utis
    )

    return [
        d_gap,
        d_eul,
        d_gyeong,
        d_shin,
        d_delta_lcdm,
        d_delta_utis
    ]


# ====================================================================
# Scan setup
# ====================================================================

S0_values = np.linspace(0.01, 0.20, 20)

tau_span = (0, 8*np.pi)
tau_eval = np.linspace(tau_span[0], tau_span[1], 4000)

Y0 = [
    0.25, 0.25, 0.25, 0.25,
    1.0, 1.0
]

suppression_results = []

# ====================================================================
# Run scan
# ====================================================================

for S0 in S0_values:

    sol = solve_ivp(
        lambda t, y: utis_growth_model_soft(
            t,
            y,
            C_growth=0.3,
            S0=S0
        ),
        tau_span,
        Y0,
        t_eval=tau_eval,
        method="Radau",
        rtol=1e-10,
        atol=1e-12
    )

    delta_lcdm_final = sol.y[4, -1]
    delta_utis_final = sol.y[5, -1]

    suppression = (
        1.0
        - delta_utis_final / delta_lcdm_final
    ) * 100.0

    suppression_results.append(suppression)

    print("------------------------------------------------")
    print(f"S0 = {S0:.3f}")
    print(f"LCDM final growth = {delta_lcdm_final:.4e}")
    print(f"UTIS final growth = {delta_utis_final:.4e}")
    print(f"Toy suppression   = {suppression:.2f}%")
    print("------------------------------------------------")


# ====================================================================
# Plot
# ====================================================================

plt.figure(figsize=(10,6))

plt.plot(
    S0_values,
    suppression_results,
    marker='o',
    linewidth=2
)

plt.axhspan(
    5,
    10,
    alpha=0.15,
    label='Target toy suppression range (~5-10%)'
)

plt.xlabel("S0 (suppression strength)")
plt.ylabel("Toy Suppression (%)")

plt.title("UTIS Soft Growth Suppression Scan")

plt.grid(True)
plt.legend()

plt.show()


# ====================================================================
# Summary
# ====================================================================

suppression_results = np.array(suppression_results)

print("========================================")
print(" 🎯 Step P5-C Summary")
print("========================================")
print(f"Min suppression : {suppression_results.min():.2f}%")
print(f"Max suppression : {suppression_results.max():.2f}%")

mask = (
    (suppression_results >= 5.0)
    &
    (suppression_results <= 10.0)
)

if np.any(mask):

    valid_S0 = S0_values[mask]

    print("✅ 5~10% toy suppression regime found")
    print(f"Candidate S0 range: {valid_S0.min():.3f} ~ {valid_S0.max():.3f}")

else:
    print("⚠️ Target suppression regime not found")

print("========================================")
print("※ 아직 toy growth proxy 단계입니다.")
print("※ 실제 σ8 / P(k) / CLASS 결과는 아닙니다.")
print("========================================")
```

```python
from pathlib import Path

CLASS_DIR = Path("/content/drive/MyDrive/UTIS_project/class_public_utis_dev")
SRC = CLASS_DIR / "source" / "utis.c"

utis_c = r'''#include "utis.h"
#include <math.h>

#ifndef _UTIS_PI_
#define _UTIS_PI_ 3.14159265358979323846
#endif

static double utis_clip_rate(double x, double xmax) {

  if (x < 0.0)
    return 0.0;

  if ((xmax > 0.0) && (x > xmax))
    return xmax;

  return x;
}

int utis_init(struct utis_parameters * putis) {

  if (putis->relay_width <= 0.0)
    putis->relay_width = 0.05;

  if (putis->min_positive <= 0.0)
    putis->min_positive = 1e-30;

  if (putis->max_rate <= 0.0)
    putis->max_rate = 1e6;

  return _SUCCESS_;
}

int utis_update_background(
  struct utis_parameters * putis,
  double tau,
  struct utis_background_state * pub
) {

  double theta;
  double W_yang, W_eum;
  double s1, s12;

  theta = fmod(putis->theta_ini + tau, 2.0 * _UTIS_PI_);

  if (theta < 0.0)
    theta += 2.0 * _UTIS_PI_;

  pub->tau = tau;
  pub->theta = theta;

  W_yang = 0.5 * (
    1.0 - tanh((theta - _UTIS_PI_) / putis->relay_width)
  );

  W_eum = 1.0 - W_yang;

  pub->W_yang = W_yang;
  pub->W_eum = W_eum;

  s1 = sin(theta);
  s12 = sin(12.0 * theta);

  pub->H_gye =
    putis->A_gye * W_yang;

  pub->T_byeong =
    putis->A_byeong * s1 * s1 * W_yang;

  pub->T_jeong =
    putis->A_jeong * W_eum;

  pub->T_im =
    putis->A_im * s1 * s1 * W_eum;

  pub->M_gi =
    s12 * s12;

  pub->Gamma_12 =
    utis_clip_rate(
      putis->gamma_12_0
      * pub->M_gi
      * pub->H_gye,
      putis->max_rate
    );

  pub->Gamma_23 =
    utis_clip_rate(
      putis->gamma_23_0
      * pub->M_gi
      * pub->T_jeong,
      putis->max_rate
    );

  pub->Gamma_34 =
    utis_clip_rate(
      putis->gamma_34_0
      * pub->T_im,
      putis->max_rate
    );

  pub->Gamma_41 =
    utis_clip_rate(
      putis->gamma_41_0
      * pub->M_gi
      * pub->H_gye,
      putis->max_rate
    );

  return _SUCCESS_;
}

int utis_update_fractions(
  struct utis_parameters * putis,
  struct utis_background_state * pub,
  struct utis_fraction_state * puf
) {

  double f0, f1, f2, f3;

  f0 = puf->f[index_utis_gap];
  f1 = puf->f[index_utis_eul];
  f2 = puf->f[index_utis_gyeong];
  f3 = puf->f[index_utis_shin];

  if (f0 < 0.0) f0 = 0.0;
  if (f1 < 0.0) f1 = 0.0;
  if (f2 < 0.0) f2 = 0.0;
  if (f3 < 0.0) f3 = 0.0;

  puf->df[index_utis_gap] =
    - pub->Gamma_12 * f0
    + pub->Gamma_41 * f3;

  puf->df[index_utis_eul] =
    + pub->Gamma_12 * f0
    - pub->Gamma_23 * f1;

  puf->df[index_utis_gyeong] =
    + pub->Gamma_23 * f1
    - pub->Gamma_34 * f2;

  puf->df[index_utis_shin] =
    + pub->Gamma_34 * f2
    - pub->Gamma_41 * f3;

  puf->total_fraction =
    f0 + f1 + f2 + f3;

  pub->clustering_fraction =
    f2 + f3;

  pub->effective_growth =
    1.0
    - putis->S0_soft
    * (1.0 - pub->clustering_fraction);

  return _SUCCESS_;
}

int utis_free(struct utis_parameters * putis) {
  return _SUCCESS_;
}
'''

SRC.write_text(utis_c)

print(f"✅ written: {SRC}")
```

```python
from pathlib import Path

UTIS_C = Path("/content/drive/MyDrive/UTIS_project/class_public_utis_dev/source/utis.c")
txt = UTIS_C.read_text()

fluid_body = r'''
/* ============================================================
 * fluid density algebraic translator
 * ============================================================ */

int utis_update_fluid_density(
  struct utis_parameters * putis,
  struct utis_background_state * pub,
  struct utis_fraction_state * puf,
  struct utis_fluid_state * pufl,
  double rho_cdm_current
){

  (void)pub;

  if (putis->has_utis == _FALSE_) {

    pufl->rho_total = rho_cdm_current;
    pufl->p_total = 0.0;

    pufl->rho[index_utis_gap] = 0.0;
    pufl->rho[index_utis_eul] = 0.0;
    pufl->rho[index_utis_gyeong] = rho_cdm_current;
    pufl->rho[index_utis_shin] = 0.0;

    pufl->p[index_utis_gap] = 0.0;
    pufl->p[index_utis_eul] = 0.0;
    pufl->p[index_utis_gyeong] = 0.0;
    pufl->p[index_utis_shin] = 0.0;

    pufl->w[index_utis_gap] = 0.0;
    pufl->w[index_utis_eul] = 0.0;
    pufl->w[index_utis_gyeong] = 0.0;
    pufl->w[index_utis_shin] = 0.0;

    pufl->cs2[index_utis_gap] = 0.0;
    pufl->cs2[index_utis_eul] = 0.0;
    pufl->cs2[index_utis_gyeong] = 0.0;
    pufl->cs2[index_utis_shin] = 0.0;

    return _SUCCESS_;
  }

  pufl->rho_total = rho_cdm_current;
  pufl->p_total = 0.0;

  pufl->rho[index_utis_gap] =
    rho_cdm_current * puf->f[index_utis_gap];

  pufl->rho[index_utis_eul] =
    rho_cdm_current * puf->f[index_utis_eul];

  pufl->rho[index_utis_gyeong] =
    rho_cdm_current * puf->f[index_utis_gyeong];

  pufl->rho[index_utis_shin] =
    rho_cdm_current * puf->f[index_utis_shin];

  pufl->w[index_utis_gap] = 0.0;
  pufl->w[index_utis_eul] = 0.0;
  pufl->w[index_utis_gyeong] = 0.0;
  pufl->w[index_utis_shin] = 0.0;

  pufl->p[index_utis_gap] = 0.0;
  pufl->p[index_utis_eul] = 0.0;
  pufl->p[index_utis_gyeong] = 0.0;
  pufl->p[index_utis_shin] = 0.0;

  pufl->cs2[index_utis_gap] = 1.0;
  pufl->cs2[index_utis_eul] = 0.01;
  pufl->cs2[index_utis_gyeong] = 0.0;
  pufl->cs2[index_utis_shin] = 0.0;

  return _SUCCESS_;
}
'''

anchor = "int utis_free("

if "int utis_update_fluid_density(" in txt:
    print("⚠️ function already exists")
else:
    if anchor not in txt:
        raise RuntimeError("anchor int utis_free not found")
    txt = txt.replace(anchor, fluid_body + "\n" + anchor)
    UTIS_C.write_text(txt)
    print("✅ utis_update_fluid_density inserted")
```
