```python
from pathlib import Path

CLASS_DIR = Path("/content/class_public_git")

utis_h = CLASS_DIR / "include" / "utis.h"
utis_c = CLASS_DIR / "source" / "utis.c"
makefile = CLASS_DIR / "Makefile"

# =========================================================
# 1. include/utis.h
# =========================================================

utis_h.write_text(r'''
#ifndef __UTIS__
#define __UTIS__

#include "background.h"

/* =========================================================
   UTIS geometry response engine
   ========================================================= */

/* phase geometry */
double utis_theta(
    struct background *pba,
    double a,
    double k
);

/* matter gravity response */
double utis_mu_m(
    struct background *pba,
    double a,
    double k
);

/* light/lensing response */
double utis_sigma_l(
    struct background *pba,
    double a,
    double k
);

/* reserved future background extensions */
double utis_F(
    struct background *pba,
    double theta
);

double utis_V(
    struct background *pba,
    double theta
);

#endif
''')

print("✅ include/utis.h created")

# =========================================================
# 2. source/utis.c
# =========================================================

utis_c.write_text(r'''
#include <math.h>

#include "utis.h"

/* =========================================================
   UTIS phase geometry
   ========================================================= */

double utis_theta(
    struct background *pba,
    double a,
    double k
){

    double A  = pba->A_theta_utis;
    double n  = pba->n_theta_utis;
    double kc = pba->k_c_override_utis;

    double k2  = k*k;
    double kc2 = kc*kc;

    return A * pow(a,n) * (k2/(k2+kc2));
}

/* =========================================================
   matter gravity response
   ========================================================= */

double utis_mu_m(
    struct background *pba,
    double a,
    double k
){

    double theta = utis_theta(pba,a,k);

    return cos(theta)*cos(theta);
}

/* =========================================================
   light/lensing response
   ========================================================= */

double utis_sigma_l(
    struct background *pba,
    double a,
    double k
){

    return 1.0;
}

/* =========================================================
   reserved future background extensions
   ========================================================= */

double utis_F(
    struct background *pba,
    double theta
){

    return 1.0;
}

double utis_V(
    struct background *pba,
    double theta
){

    return 0.0;
}
''')

print("✅ source/utis.c created")

# =========================================================
# 3. Makefile patch
# =========================================================

txt = makefile.read_text()

if "utis.o" not in txt:

    txt = txt.replace(
        "source/lensing.o",
        "source/lensing.o \\\nsource/utis.o"
    )

    makefile.write_text(txt)

    print("✅ Makefile patched")

else:
    print("ℹ️ Makefile already patched")

print("\n🎯 UTIS central engine scaffold complete")
```

```python
from pathlib import Path
import subprocess
import numpy as np

CLASS_DIR = Path("/content/class_public_git")
assert CLASS_DIR.exists(), "❌ CLASS_DIR 없음"
assert (CLASS_DIR / "class").exists(), "❌ class executable 없음. 먼저 make -j class 필요"

# =========================================================
# helper: 기존 key 전부 제거 후 하나만 다시 추가
# =========================================================

def remove_param(text, key):
    out = []
    for line in text.splitlines():
        s = line.strip()
        if s.startswith(key + " ") or s.startswith(key + "="):
            continue
        out.append(line)
    return "\n".join(out) + "\n"

def set_unique(text, key, value):
    text = remove_param(text, key)
    return text + f"\n{key} = {value}\n"

# =========================================================
# base ini: explanatory.ini에서 시작
# =========================================================

base = (CLASS_DIR / "explanatory.ini").read_text()

common = base
for key, val in [
    ("output", "mPk"),
    ("modes", "s"),
    ("gauge", "newtonian"),
    ("lensing", "no"),
    ("P_k_max_h/Mpc", "1.0"),
    ("z_pk", "0"),
]:
    common = set_unique(common, key, val)

# =========================================================
# UTIS OFF
# =========================================================

off = common
for key, val in [
    ("root", "output/utis_off_"),
    ("has_utis", "no"),
    ("A_theta_utis", "0."),
    ("n_theta_utis", "0.5"),
    ("k_c_override_utis", "0.2"),
]:
    off = set_unique(off, key, val)

# =========================================================
# UTIS ON
# =========================================================

on = common
for key, val in [
    ("root", "output/utis_on_"),
    ("has_utis", "yes"),
    ("A_theta_utis", "1.25"),
    ("n_theta_utis", "0.5"),
    ("k_c_override_utis", "0.2"),
]:
    on = set_unique(on, key, val)

(CLASS_DIR / "utis_off.ini").write_text(off)
(CLASS_DIR / "utis_on.ini").write_text(on)

print("✅ utis_off.ini written")
print("✅ utis_on.ini written")

# =========================================================
# run CLASS
# =========================================================

for ini in ["utis_off.ini", "utis_on.ini"]:
    print("\n🚀 RUN", ini)
    res = subprocess.run(
        ["./class", ini],
        cwd=CLASS_DIR,
        text=True,
        capture_output=True
    )
    print("returncode =", res.returncode)

    if res.returncode != 0:
        print("\nSTDOUT:")
        print(res.stdout[-5000:])
        print("\nSTDERR:")
        print(res.stderr[-5000:])
        raise RuntimeError(f"❌ CLASS failed: {ini}")

    print(res.stdout[-1000:])

# =========================================================
# locate P(k) files
# =========================================================

out_dir = CLASS_DIR / "output"

def find_pk(prefix):
    cands = sorted(set(
        list(out_dir.glob(prefix + "*pk*.dat")) +
        list(out_dir.glob(prefix + "*mPk*.dat"))
    ))
    if not cands:
        print("Available files:")
        for p in sorted(out_dir.glob(prefix + "*")):
            print(" -", p.name)
        raise FileNotFoundError(f"❌ no P(k) file for prefix={prefix}")
    return cands[0]

pk_off = find_pk("utis_off_")
pk_on = find_pk("utis_on_")

print("\n✅ pk_off =", pk_off.name)
print("✅ pk_on  =", pk_on.name)

# =========================================================
# load / ratio
# =========================================================

d_off = np.loadtxt(pk_off)
d_on = np.loadtxt(pk_on)

k_off, P_off = d_off[:, 0], d_off[:, 1]
k_on, P_on = d_on[:, 0], d_on[:, 1]

if len(k_off) != len(k_on) or np.max(np.abs(k_off - k_on)) > 1e-14:
    print("ℹ️ interpolating ON grid")
    P_on_i = np.interp(k_off, k_on, P_on)
else:
    P_on_i = P_on

ratio = P_on_i / P_off

# =========================================================
# diagnostics
# =========================================================

print("\n" + "="*80)
print("📊 UTIS P(k) ratio diagnostic")
print("="*80)

print("ratio min      =", ratio.min())
print("ratio max      =", ratio.max())
print("ratio mean     =", ratio.mean())
print("max |ratio-1|  =", np.max(np.abs(ratio - 1.0)))

print("\n📌 sample points")
step = max(1, len(k_off)//12)

for kk, rr in zip(k_off[::step], ratio[::step]):
    print(f"k={kk:.6e}   ratio={rr:.10f}")

if np.max(np.abs(ratio - 1.0)) < 1e-8:
    print("\n⚠️ ratio ≈ 1")
    print("UTIS hook이 아직 P(k)에 실질적으로 전달되지 않았을 가능성.")
else:
    print("\n✅ ratio deviates from 1")
    print("UTIS metric_euler hook이 실제 evolution에 영향을 준 것.")
```

```python
# ============================================================
# UTIS GLOBAL LIKELIHOOD v2
# SN + H(z) + BAO(DM,DH) + RSD
# ============================================================

from classy import Class
import pandas as pd
import numpy as np

# ============================================================
# datasets
# ============================================================

print("="*80)
print("LOAD DATASETS")
print("="*80)

RSD = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/RSD_clean.csv")
HZ  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/CC_clean.csv")
BAO = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/BAO_clean.csv")
SN  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/SN_pantheon_clean.csv")

print("RSD =", len(RSD), RSD.columns.tolist())
print("HZ  =", len(HZ), HZ.columns.tolist())
print("BAO =", len(BAO), BAO.columns.tolist())
print("SN  =", len(SN), SN.columns.tolist())

# ============================================================
# baseline cosmology
# ============================================================

base_pars = {
    'output': 'mPk,lCl,pCl',
    'lensing': 'yes',

    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,
}

utis_best = {
    'has_utis': 'yes',

    'utis_s0': 0.12,
    'utis_sigma0': 0.02,
    'utis_kc': 0.05,

    'A_theta_utis': 0.0,
    'n_theta_utis': 0.8,
    'k_c_override_utis': 0.05,
}

c = 299792.458

# ============================================================
# chi2 helper
# ============================================================

def compute_all_chi2(cosmo):

    # --------------------------------------------------------
    # RSD: f sigma8
    # --------------------------------------------------------

    z = RSD['z'].values
    obs = RSD['fs8'].values
    err = RSD['err'].values

    th = []

    for zz in z:
        f = cosmo.scale_independent_growth_factor_f(zz)
        s8 = cosmo.sigma(8, zz)
        th.append(f*s8)

    th = np.array(th)

    chi2_rsd = np.sum(((obs-th)/err)**2)

    # --------------------------------------------------------
    # H(z): cosmic chronometers
    # CLASS Hubble(z) is 1/Mpc, multiply by c -> km/s/Mpc
    # --------------------------------------------------------

    z = HZ['z'].values
    obs = HZ['H'].values
    err = HZ['err'].values

    th = []

    for zz in z:
        H = cosmo.Hubble(zz) * c
        th.append(H)

    th = np.array(th)

    chi2_hz = np.sum(((obs-th)/err)**2)

    # --------------------------------------------------------
    # SN: Pantheon-like, analytic M offset
    # --------------------------------------------------------

    z = SN['z'].values
    mu_obs = SN['mu'].values
    mu_err = SN['mu_err'].values

    mu_th = []

    for zz in z:
        dL = cosmo.luminosity_distance(zz)  # Mpc
        mu = 5*np.log10(dL) + 25
        mu_th.append(mu)

    mu_th = np.array(mu_th)

    delta_mu = mu_obs - mu_th
    w = 1.0 / mu_err**2
    M_offset = np.sum(w * delta_mu) / np.sum(w)

    chi2_sn = np.sum(((delta_mu - M_offset)/mu_err)**2)

    # --------------------------------------------------------
    # BAO: DM/rd and DH/rd diagonal approximation
    # --------------------------------------------------------

    chi2_bao = 0.0
    rd = cosmo.rs_drag()

    for _, row in BAO.iterrows():

        zz = row['z']

        DA = cosmo.angular_distance(zz)       # Mpc
        DM = (1.0 + zz) * DA                  # transverse comoving distance
        H  = cosmo.Hubble(zz) * c             # km/s/Mpc
        DH = c / H                            # Mpc

        if np.isfinite(row.get('DM_over_rd', np.nan)) and np.isfinite(row.get('DM_over_rd_err', np.nan)):
            th = DM / rd
            obs = row['DM_over_rd']
            err = row['DM_over_rd_err']
            chi2_bao += ((obs-th)/err)**2

        if np.isfinite(row.get('DH_over_rd', np.nan)) and np.isfinite(row.get('DH_over_rd_err', np.nan)):
            th = DH / rd
            obs = row['DH_over_rd']
            err = row['DH_over_rd_err']
            chi2_bao += ((obs-th)/err)**2

    total = chi2_rsd + chi2_hz + chi2_sn + chi2_bao

    return {
        'chi2_rsd': chi2_rsd,
        'chi2_hz': chi2_hz,
        'chi2_sn': chi2_sn,
        'chi2_bao': chi2_bao,
        'chi2_total': total,
        'SN_M_offset': M_offset,
    }

# ============================================================
# LCDM
# ============================================================

print("\n" + "="*80)
print("LCDM")
print("="*80)

lcdm = Class()
lcdm.set(base_pars)
lcdm.compute()

lcdm_sigma8 = lcdm.sigma8()
lcdm_res = compute_all_chi2(lcdm)

print("sigma8 =", lcdm_sigma8)
for k,v in lcdm_res.items():
    print(k, "=", v)

# ============================================================
# UTIS
# ============================================================

print("\n" + "="*80)
print("UTIS")
print("="*80)

pars_utis = base_pars.copy()
pars_utis.update(utis_best)

utis = Class()
utis.set(pars_utis)
utis.compute()

utis_sigma8 = utis.sigma8()
utis_res = compute_all_chi2(utis)

print("sigma8 =", utis_sigma8)
for k,v in utis_res.items():
    print(k, "=", v)

# ============================================================
# comparison
# ============================================================

rows = []

for key in ['chi2_rsd','chi2_hz','chi2_sn','chi2_bao','chi2_total']:

    rows.append({
        'sector': key,
        'LCDM': lcdm_res[key],
        'UTIS': utis_res[key],
        'delta': utis_res[key] - lcdm_res[key],
    })

tbl = pd.DataFrame(rows)

print("\n" + "="*80)
print("COMPARISON")
print("="*80)

display(tbl)

print("\n" + "="*80)
print("FINAL")
print("="*80)

print("sigma8_LCDM =", lcdm_sigma8)
print("sigma8_UTIS =", utis_sigma8)
print("sigma8_ratio =", utis_sigma8/lcdm_sigma8)
print("delta_total =", utis_res['chi2_total'] - lcdm_res['chi2_total'])

lcdm.struct_cleanup()
lcdm.empty()

utis.struct_cleanup()
utis.empty()
```

```python
# ============================================================
# UTIS GLOBAL GRID SCAN
# SN + HZ + BAO + RSD
# ============================================================

from classy import Class
import pandas as pd
import numpy as np
from itertools import product

# ============================================================
# datasets
# ============================================================

RSD = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/RSD_clean.csv")
HZ  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/CC_clean.csv")
BAO = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/BAO_clean.csv")
SN  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/SN_pantheon_clean.csv")

# ============================================================
# baseline cosmology
# ============================================================

base_pars = {

    'output': 'mPk,lCl,pCl',
    'lensing': 'yes',

    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,
}

c = 299792.458

# ============================================================
# chi2 helper
# ============================================================

def compute_all_chi2(cosmo):

    # --------------------------------------------------------
    # RSD
    # --------------------------------------------------------

    z = RSD['z'].values
    obs = RSD['fs8'].values
    err = RSD['err'].values

    th = []

    for zz in z:

        f = cosmo.scale_independent_growth_factor_f(zz)
        s8 = cosmo.sigma(8, zz)

        th.append(f*s8)

    th = np.array(th)

    chi2_rsd = np.sum(((obs-th)/err)**2)

    # --------------------------------------------------------
    # H(z)
    # --------------------------------------------------------

    z = HZ['z'].values
    obs = HZ['H'].values
    err = HZ['err'].values

    th = []

    for zz in z:

        H = cosmo.Hubble(zz) * c
        th.append(H)

    th = np.array(th)

    chi2_hz = np.sum(((obs-th)/err)**2)

    # --------------------------------------------------------
    # SN
    # --------------------------------------------------------

    z = SN['z'].values
    mu_obs = SN['mu'].values
    mu_err = SN['mu_err'].values

    mu_th = []

    for zz in z:

        dL = cosmo.luminosity_distance(zz)

        mu = 5*np.log10(dL) + 25

        mu_th.append(mu)

    mu_th = np.array(mu_th)

    delta_mu = mu_obs - mu_th

    w = 1.0 / mu_err**2

    M_offset = np.sum(w * delta_mu) / np.sum(w)

    chi2_sn = np.sum(
        ((delta_mu - M_offset)/mu_err)**2
    )

    # --------------------------------------------------------
    # BAO
    # --------------------------------------------------------

    chi2_bao = 0.0

    rd = cosmo.rs_drag()

    for _, row in BAO.iterrows():

        zz = row['z']

        DA = cosmo.angular_distance(zz)

        DM = (1.0 + zz) * DA

        H = cosmo.Hubble(zz) * c

        DH = c / H

        # DM/rd
        if np.isfinite(row['DM_over_rd']):

            obs = row['DM_over_rd']
            err = row['DM_over_rd_err']

            th = DM / rd

            chi2_bao += ((obs-th)/err)**2

        # DH/rd
        if np.isfinite(row['DH_over_rd']):

            obs = row['DH_over_rd']
            err = row['DH_over_rd_err']

            th = DH / rd

            chi2_bao += ((obs-th)/err)**2

    total = (
        chi2_rsd
        + chi2_hz
        + chi2_sn
        + chi2_bao
    )

    return {
        'chi2_rsd': chi2_rsd,
        'chi2_hz': chi2_hz,
        'chi2_sn': chi2_sn,
        'chi2_bao': chi2_bao,
        'chi2_total': total,
    }

# ============================================================
# LCDM reference
# ============================================================

print("="*80)
print("LCDM reference")
print("="*80)

lcdm = Class()
lcdm.set(base_pars)
lcdm.compute()

lcdm_sigma8 = lcdm.sigma8()

lcdm_res = compute_all_chi2(lcdm)

print("sigma8_LCDM =", lcdm_sigma8)
print("chi2_LCDM =", lcdm_res['chi2_total'])

lcdm.struct_cleanup()
lcdm.empty()

# ============================================================
# scan ranges
# ============================================================

s0_list = [
    0.08,
    0.10,
    0.12,
    0.14,
]

sigma0_list = [
    0.00,
    0.01,
    0.02,
    0.03,
]

kc_list = [
    0.04,
    0.05,
    0.06,
    0.08,
]

n_list = [
    0.6,
    0.8,
    1.0,
    1.2,
]

grid = list(product(
    s0_list,
    sigma0_list,
    kc_list,
    n_list,
))

print("\n")
print("="*80)
print("TOTAL GRID =", len(grid))
print("="*80)

# ============================================================
# scan
# ============================================================

rows = []

for i, (s0, sigma0, kc, nn) in enumerate(grid):

    print("\n")
    print("="*80)
    print(f"[{i+1}/{len(grid)}]")
    print("="*80)

    pars = base_pars.copy()

    pars.update({

        'has_utis': 'yes',

        'utis_s0': s0,
        'utis_sigma0': sigma0,
        'utis_kc': kc,

        'A_theta_utis': 0.0,
        'n_theta_utis': nn,
        'k_c_override_utis': kc,
    })

    try:

        cosmo = Class()

        cosmo.set(pars)
        cosmo.compute()

        sigma8 = cosmo.sigma8()

        res = compute_all_chi2(cosmo)

        print("PASS")
        print("sigma8 =", sigma8)
        print("chi2 =", res['chi2_total'])

        rows.append({

            'status': 'PASS',

            's0': s0,
            'sigma0': sigma0,
            'kc': kc,
            'n_theta': nn,

            'sigma8': sigma8,
            'sigma8_ratio': sigma8 / lcdm_sigma8,

            'chi2_lcdm': lcdm_res['chi2_total'],
            'chi2_utis': res['chi2_total'],
            'delta_chi2': (
                res['chi2_total']
                - lcdm_res['chi2_total']
            ),

            'chi2_rsd': res['chi2_rsd'],
            'chi2_hz': res['chi2_hz'],
            'chi2_sn': res['chi2_sn'],
            'chi2_bao': res['chi2_bao'],
        })

        cosmo.struct_cleanup()
        cosmo.empty()

    except Exception as e:

        print("FAIL")
        print(str(e)[:300])

        rows.append({

            'status': 'FAIL',

            's0': s0,
            'sigma0': sigma0,
            'kc': kc,
            'n_theta': nn,

            'error': str(e),
        })

# ============================================================
# summary
# ============================================================

tbl = pd.DataFrame(rows)

good = (
    tbl[tbl['status']=='PASS']
    .sort_values('delta_chi2')
)

print("\n")
print("="*80)
print("TOP MODELS")
print("="*80)

display(good.head(30))

save_path = "/content/utis_global_scan.csv"

tbl.to_csv(save_path, index=False)

print("\n")
print("saved:", save_path)
```

```python
# ============================================================
# UTIS Planck-lite spectra audit
# TT / TE / EE / PP residual + pseudo chi2
# ============================================================

from classy import Class
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# parameters
# ============================================================

pars_lcdm = {
    'output': 'tCl,pCl,lCl,mPk',
    'lensing': 'yes',

    'l_max_scalars': 2500,
    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,
}

pars_utis = pars_lcdm.copy()
pars_utis.update({
    'has_utis': 'yes',

    # current dense-scan best region
    'utis_s0': 0.18,
    'utis_sigma0': 0.02,
    'utis_kc': 0.025,

    'A_theta_utis': 0.0,
    'n_theta_utis': 0.6,
    'k_c_override_utis': 0.025,
})

# ============================================================
# compute
# ============================================================

print("="*80)
print("Compute LCDM")
print("="*80)

lcdm = Class()
lcdm.set(pars_lcdm)
lcdm.compute()

cl_lcdm_raw = lcdm.raw_cl(2500)
cl_lcdm_len = lcdm.lensed_cl(2500)

sigma8_lcdm = lcdm.sigma8()
print("sigma8_LCDM =", sigma8_lcdm)

print("\n" + "="*80)
print("Compute UTIS")
print("="*80)

utis = Class()
utis.set(pars_utis)
utis.compute()

cl_utis_raw = utis.raw_cl(2500)
cl_utis_len = utis.lensed_cl(2500)

sigma8_utis = utis.sigma8()
print("sigma8_UTIS =", sigma8_utis)
print("sigma8_ratio =", sigma8_utis/sigma8_lcdm)

# ============================================================
# helpers
# ============================================================

def safe_ratio_stats(ell, a, b, name, rel_floor=1e-5, lmin=2, lmax=2500):
    """
    Computes residual ratio safely, excluding zero-crossing tiny denominator regions.
    rel_floor excludes |LCDM| < rel_floor * max(|LCDM|).
    """
    ell = np.asarray(ell)
    a = np.asarray(a)
    b = np.asarray(b)

    mask = (
        (ell >= lmin)
        & (ell <= lmax)
        & np.isfinite(a)
        & np.isfinite(b)
    )

    denom_scale = np.nanmax(np.abs(a[mask]))
    mask = mask & (np.abs(a) > rel_floor * denom_scale)

    ratio = b[mask] / a[mask]
    resid = ratio - 1.0

    return {
        "name": name,
        "valid": int(np.sum(mask)),
        "max_abs_resid": float(np.max(np.abs(resid))) if np.sum(mask) else np.nan,
        "mean_abs_resid": float(np.mean(np.abs(resid))) if np.sum(mask) else np.nan,
        "rms_resid": float(np.sqrt(np.mean(resid**2))) if np.sum(mask) else np.nan,
        "min_ratio": float(np.min(ratio)) if np.sum(mask) else np.nan,
        "max_ratio": float(np.max(ratio)) if np.sum(mask) else np.nan,
    }, mask, resid

def pseudo_chi2_fractional(ell, cl_ref, cl_alt, f_sky=0.7, noise_frac=0.01, lmin=30, lmax=2000, rel_floor=1e-5):
    """
    Planck-lite pseudo chi2.
    Uses fractional residual divided by approximate cosmic variance + noise floor.

    sigma_frac^2 ≈ 2/[(2l+1)f_sky] + noise_frac^2
    """
    ell = np.asarray(ell)
    cl_ref = np.asarray(cl_ref)
    cl_alt = np.asarray(cl_alt)

    mask = (
        (ell >= lmin)
        & (ell <= lmax)
        & np.isfinite(cl_ref)
        & np.isfinite(cl_alt)
    )

    scale = np.nanmax(np.abs(cl_ref[mask]))
    mask = mask & (np.abs(cl_ref) > rel_floor * scale)

    frac = cl_alt[mask]/cl_ref[mask] - 1.0
    sigma_frac = np.sqrt(2.0/((2*ell[mask]+1)*f_sky) + noise_frac**2)

    chi2 = np.sum((frac/sigma_frac)**2)

    return float(chi2), int(np.sum(mask)), float(chi2/max(np.sum(mask),1))

# ============================================================
# residual audits
# ============================================================

ell_raw = cl_lcdm_raw['ell']
ell_len = cl_lcdm_len['ell']

summary = []

# RAW primary spectra
for key, label in [
    ('tt', 'raw_TT'),
    ('ee', 'raw_EE'),
    ('te', 'raw_TE'),
    ('pp', 'raw_PP'),
    ('tp', 'raw_TP'),
]:
    stats, _, _ = safe_ratio_stats(
        ell_raw,
        cl_lcdm_raw[key],
        cl_utis_raw[key],
        label,
        rel_floor=1e-4 if key in ['te','tp'] else 1e-6,
        lmin=2,
        lmax=2500,
    )
    summary.append(stats)

# LENSED spectra
for key, label in [
    ('tt', 'lensed_TT'),
    ('ee', 'lensed_EE'),
    ('te', 'lensed_TE'),
    ('pp', 'lensed_PP'),
    ('tp', 'lensed_TP'),
]:
    stats, _, _ = safe_ratio_stats(
        ell_len,
        cl_lcdm_len[key],
        cl_utis_len[key],
        label,
        rel_floor=1e-4 if key in ['te','tp'] else 1e-6,
        lmin=2,
        lmax=2500,
    )
    summary.append(stats)

summary = pd.DataFrame(summary)

print("\n" + "="*80)
print("RESIDUAL SUMMARY")
print("="*80)
display(summary)

# ============================================================
# pseudo chi2
# ============================================================

pseudo_rows = []

settings = {
    "TT": {"key": "tt", "noise_frac": 0.005, "lmin": 30, "lmax": 2000, "rel_floor": 1e-6},
    "EE": {"key": "ee", "noise_frac": 0.015, "lmin": 30, "lmax": 2000, "rel_floor": 1e-6},
    "TE": {"key": "te", "noise_frac": 0.020, "lmin": 30, "lmax": 2000, "rel_floor": 1e-4},
    "PP": {"key": "pp", "noise_frac": 0.030, "lmin": 8,  "lmax": 1000, "rel_floor": 1e-6},
}

for obs, cfg in settings.items():
    chi2, n, red = pseudo_chi2_fractional(
        ell_len,
        cl_lcdm_len[cfg["key"]],
        cl_utis_len[cfg["key"]],
        f_sky=0.7,
        noise_frac=cfg["noise_frac"],
        lmin=cfg["lmin"],
        lmax=cfg["lmax"],
        rel_floor=cfg["rel_floor"],
    )

    pseudo_rows.append({
        "observable": obs,
        "pseudo_chi2": chi2,
        "n_eff": n,
        "pseudo_chi2_per_mode": red,
        "noise_frac_assumed": cfg["noise_frac"],
        "lmin": cfg["lmin"],
        "lmax": cfg["lmax"],
    })

pseudo = pd.DataFrame(pseudo_rows)

print("\n" + "="*80)
print("PLANCK-LITE PSEUDO CHI2")
print("="*80)
display(pseudo)

print("\n" + "="*80)
print("TOTAL PSEUDO CHI2")
print("="*80)
print("pseudo_chi2_total =", pseudo["pseudo_chi2"].sum())
print("pseudo_chi2_per_mode_mean =", pseudo["pseudo_chi2"].sum()/pseudo["n_eff"].sum())

# ============================================================
# window audit for PP
# ============================================================

ells = ell_len[2:]
pp_lcdm = cl_lcdm_len['pp'][2:]
pp_utis = cl_utis_len['pp'][2:]

ratio_pp = pp_utis / pp_lcdm
delta_pp = ratio_pp - 1.0

windows = {
    "L_8_40": (8,40),
    "L_40_100": (40,100),
    "L_100_400": (100,400),
    "L_400_1000": (400,1000),
}

win_rows = []

for name, (Lmin, Lmax) in windows.items():
    m = (ells >= Lmin) & (ells <= Lmax)
    win_rows.append({
        "window": name,
        "max_abs_PP_resid": float(np.max(np.abs(delta_pp[m]))),
        "mean_abs_PP_resid": float(np.mean(np.abs(delta_pp[m]))),
    })

win_tbl = pd.DataFrame(win_rows)

print("\n" + "="*80)
print("PP WINDOW AUDIT")
print("="*80)
display(win_tbl)

# ============================================================
# plots
# ============================================================

def plot_residual(ell, ref, alt, title, ylabel, rel_floor=1e-6):
    ell = np.asarray(ell)
    ref = np.asarray(ref)
    alt = np.asarray(alt)

    mask = (ell >= 2) & np.isfinite(ref) & np.isfinite(alt)
    scale = np.nanmax(np.abs(ref[mask]))
    mask = mask & (np.abs(ref) > rel_floor * scale)

    plt.figure(figsize=(10,5))
    plt.plot(ell[mask], alt[mask]/ref[mask] - 1.0)
    plt.axhline(0, linestyle='--')
    plt.xscale('log')
    plt.xlabel(r"$\ell$ or $L$")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_residual(
    ell_len,
    cl_lcdm_len['tt'],
    cl_utis_len['tt'],
    "Lensed TT residual",
    r"$C_\ell^{TT,\mathrm{UTIS}}/C_\ell^{TT,\Lambda CDM}-1$",
)

plot_residual(
    ell_len,
    cl_lcdm_len['ee'],
    cl_utis_len['ee'],
    "Lensed EE residual",
    r"$C_\ell^{EE,\mathrm{UTIS}}/C_\ell^{EE,\Lambda CDM}-1$",
)

plot_residual(
    ell_len,
    cl_lcdm_len['te'],
    cl_utis_len['te'],
    "Lensed TE residual, zero-crossing safe",
    r"$C_\ell^{TE,\mathrm{UTIS}}/C_\ell^{TE,\Lambda CDM}-1$",
    rel_floor=1e-4,
)

plot_residual(
    ell_len,
    cl_lcdm_len['pp'],
    cl_utis_len['pp'],
    "Lensing PP residual",
    r"$C_L^{\phi\phi,\mathrm{UTIS}}/C_L^{\phi\phi,\Lambda CDM}-1$",
)

# ============================================================
# save
# ============================================================

summary.to_csv("/content/utis_planck_lite_residual_summary.csv", index=False)
pseudo.to_csv("/content/utis_planck_lite_pseudo_chi2.csv", index=False)
win_tbl.to_csv("/content/utis_planck_lite_pp_windows.csv", index=False)

print("\nSaved:")
print("/content/utis_planck_lite_residual_summary.csv")
print("/content/utis_planck_lite_pseudo_chi2.csv")
print("/content/utis_planck_lite_pp_windows.csv")

# ============================================================
# cleanup
# ============================================================

lcdm.struct_cleanup()
lcdm.empty()

utis.struct_cleanup()
utis.empty()
```

```python
# ============================================================
# UTIS SAFE vs AGGRESSIVE benchmark comparison
# paper-ready benchmark table
# ============================================================

from classy import Class
import numpy as np
import pandas as pd

# ============================================================
# base LCDM
# ============================================================

base = {
    'output': 'tCl,pCl,lCl,mPk',
    'lensing': 'yes',

    'l_max_scalars': 2500,
    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,
}

# ============================================================
# benchmark models
# ============================================================

models = {

    "LCDM": {},

    # conservative / publication-safe
    "UTIS_safe": {
        'has_utis': 'yes',

        'utis_s0': 0.14,
        'utis_sigma0': 0.02,
        'utis_kc': 0.04,

        'A_theta_utis': 0.0,
        'n_theta_utis': 0.6,
        'k_c_override_utis': 0.04,
    },

    # strong growth suppression
    "UTIS_aggressive": {
        'has_utis': 'yes',

        'utis_s0': 0.18,
        'utis_sigma0': 0.02,
        'utis_kc': 0.025,

        'A_theta_utis': 0.0,
        'n_theta_utis': 0.6,
        'k_c_override_utis': 0.025,
    },
}

# ============================================================
# helper
# ============================================================

def pp_residual_stats(cl_ref, cl_alt):

    ell = cl_ref['ell'][2:]

    pp_ref = cl_ref['pp'][2:]
    pp_alt = cl_alt['pp'][2:]

    ratio = pp_alt / pp_ref
    resid = ratio - 1.0

    return {
        "pp_max_abs": float(np.max(np.abs(resid))),
        "pp_mean_abs": float(np.mean(np.abs(resid))),
    }

def lensed_residual_max(cl_ref, cl_alt, key, rel_floor=1e-6):

    ref = cl_ref[key]
    alt = cl_alt[key]

    scale = np.max(np.abs(ref))

    mask = np.abs(ref) > rel_floor * scale

    resid = alt[mask]/ref[mask] - 1.0

    return float(np.max(np.abs(resid)))

# ============================================================
# compute LCDM first
# ============================================================

print("="*80)
print("LCDM baseline")
print("="*80)

lcdm = Class()
lcdm.set(base)
lcdm.compute()

cl_lcdm = lcdm.lensed_cl(2500)

sigma8_lcdm = lcdm.sigma8()

print("sigma8_LCDM =", sigma8_lcdm)

# ============================================================
# benchmark loop
# ============================================================

rows = []

for name, extra in models.items():

    print("\n" + "="*80)
    print(name)
    print("="*80)

    pars = base.copy()
    pars.update(extra)

    cosmo = Class()
    cosmo.set(pars)
    cosmo.compute()

    sigma8 = cosmo.sigma8()

    cl = cosmo.lensed_cl(2500)

    # residuals
    tt_max = lensed_residual_max(cl_lcdm, cl, 'tt')
    ee_max = lensed_residual_max(cl_lcdm, cl, 'ee')
    te_max = lensed_residual_max(cl_lcdm, cl, 'te', rel_floor=1e-4)

    pp_stats = pp_residual_stats(cl_lcdm, cl)

    row = {
        "model": name,

        "sigma8": sigma8,
        "sigma8_ratio": sigma8/sigma8_lcdm,

        "TT_max_resid": tt_max,
        "EE_max_resid": ee_max,
        "TE_max_resid": te_max,

        "PP_max_resid": pp_stats["pp_max_abs"],
        "PP_mean_resid": pp_stats["pp_mean_abs"],
    }

    # UTIS-only params
    row["s0"] = extra.get("utis_s0", 0.0)
    row["sigma0"] = extra.get("utis_sigma0", 0.0)
    row["kc"] = extra.get("utis_kc", 0.0)
    row["n_theta"] = extra.get("n_theta_utis", 0.0)

    rows.append(row)

    print("sigma8 =", sigma8)
    print("sigma8_ratio =", sigma8/sigma8_lcdm)
    print("TT_max =", tt_max)
    print("EE_max =", ee_max)
    print("TE_max =", te_max)
    print("PP_max =", pp_stats["pp_max_abs"])

    cosmo.struct_cleanup()
    cosmo.empty()

# ============================================================
# summary
# ============================================================

tbl = pd.DataFrame(rows)

print("\n" + "="*80)
print("BENCHMARK SUMMARY")
print("="*80)

display(tbl)

# ============================================================
# interpretation labels
# ============================================================

labels = []

for _, r in tbl.iterrows():

    if r["model"] == "LCDM":
        labels.append("baseline")
        continue

    if r["PP_max_resid"] < 0.01:
        lensing = "ultra-safe"
    elif r["PP_max_resid"] < 0.02:
        lensing = "safe"
    else:
        lensing = "aggressive"

    if r["sigma8_ratio"] > 0.95:
        growth = "mild"
    elif r["sigma8_ratio"] > 0.92:
        growth = "moderate"
    else:
        growth = "strong"

    labels.append(f"{growth} growth suppression + {lensing} lensing")

tbl["interpretation"] = labels

print("\n" + "="*80)
print("INTERPRETATION")
print("="*80)

display(tbl)

# ============================================================
# save
# ============================================================

out = "/content/utis_safe_vs_aggressive.csv"

tbl.to_csv(out, index=False)

print("\nSaved:")
print(out)

# ============================================================
# cleanup
# ============================================================

lcdm.struct_cleanup()
lcdm.empty()
```

```python
# ============================================================
# UTIS EXTENDED MCMC
# params: s0, kc, n_theta, ln10As
# fixed: sigma0 = 0.02
# prior: s0 up to 0.35
# likelihood: SN + HZ + BAO + RSD
# ============================================================

!pip -q install emcee corner

from classy import Class
import numpy as np
import pandas as pd
import emcee
import corner
import matplotlib.pyplot as plt

# ============================================================
# Load datasets
# ============================================================

RSD = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/RSD_clean.csv")
HZ  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/CC_clean.csv")
BAO = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/BAO_clean.csv")
SN  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/SN_pantheon_clean.csv")

c = 299792.458

# ============================================================
# Base cosmology
# ============================================================

base = {
    "output": "mPk,lCl,pCl",
    "lensing": "yes",
    "P_k_max_h/Mpc": 10.0,
    "z_max_pk": 5.0,

    "h": 0.6736,
    "omega_b": 0.02237,
    "omega_cdm": 0.1200,

    "n_s": 0.9649,
    "tau_reio": 0.0544,
}

# ============================================================
# Convert ln(10^10 A_s) to A_s
# ============================================================

def As_from_ln10As(ln10As):
    return np.exp(ln10As) / 1e10

# ============================================================
# chi2
# ============================================================

def compute_chi2(s0, kc, ntheta, ln10As, sigma0=0.02):

    pars = base.copy()

    pars.update({
        "A_s": float(As_from_ln10As(ln10As)),

        "has_utis": "yes",
        "utis_s0": float(s0),
        "utis_sigma0": float(sigma0),
        "utis_kc": float(kc),

        "A_theta_utis": 0.0,
        "n_theta_utis": float(ntheta),
        "k_c_override_utis": float(kc),
    })

    cosmo = None

    try:
        cosmo = Class()
        cosmo.set(pars)
        cosmo.compute()

        # RSD
        fs8_th = []
        for z in RSD["z"].values:
            f = cosmo.scale_independent_growth_factor_f(z)
            s8 = cosmo.sigma(8, z)
            fs8_th.append(f*s8)

        fs8_th = np.array(fs8_th)
        chi2_rsd = np.sum(((RSD["fs8"].values - fs8_th) / RSD["err"].values)**2)

        # H(z)
        H_th = np.array([cosmo.Hubble(z) * c for z in HZ["z"].values])
        chi2_hz = np.sum(((HZ["H"].values - H_th) / HZ["err"].values)**2)

        # SN with analytic M offset
        mu_th = []
        for z in SN["z"].values:
            dL = cosmo.luminosity_distance(z)
            mu_th.append(5*np.log10(dL) + 25)

        mu_th = np.array(mu_th)
        delta_mu = SN["mu"].values - mu_th
        w = 1.0 / SN["mu_err"].values**2
        M_offset = np.sum(w * delta_mu) / np.sum(w)
        chi2_sn = np.sum(((delta_mu - M_offset) / SN["mu_err"].values)**2)

        # BAO diagonal approximation
        rd = cosmo.rs_drag()
        chi2_bao = 0.0

        for _, row in BAO.iterrows():
            z = row["z"]
            DA = cosmo.angular_distance(z)
            DM = (1.0 + z) * DA
            H = cosmo.Hubble(z) * c
            DH = c / H

            if np.isfinite(row["DM_over_rd"]) and np.isfinite(row["DM_over_rd_err"]):
                chi2_bao += ((row["DM_over_rd"] - DM/rd) / row["DM_over_rd_err"])**2

            if np.isfinite(row["DH_over_rd"]) and np.isfinite(row["DH_over_rd_err"]):
                chi2_bao += ((row["DH_over_rd"] - DH/rd) / row["DH_over_rd_err"])**2

        sigma8 = cosmo.sigma8()
        total = chi2_rsd + chi2_hz + chi2_sn + chi2_bao

        return total, chi2_rsd, chi2_hz, chi2_sn, chi2_bao, sigma8

    except Exception:
        return np.inf, np.inf, np.inf, np.inf, np.inf, np.nan

    finally:
        if cosmo is not None:
            try:
                cosmo.struct_cleanup()
                cosmo.empty()
            except Exception:
                pass

# ============================================================
# Prior
# ============================================================

def log_prior(theta):

    s0, kc, ntheta, ln10As = theta

    if not (0.00 <= s0 <= 0.35):
        return -np.inf

    if not (0.010 <= kc <= 0.12):
        return -np.inf

    if not (0.20 <= ntheta <= 1.50):
        return -np.inf

    # Planck-like broad prior around ln(10^10 As) ~ 3.04
    if not (2.70 <= ln10As <= 3.30):
        return -np.inf

    return 0.0

cache = {}

def log_prob(theta):

    lp = log_prior(theta)

    if not np.isfinite(lp):
        return -np.inf

    key = tuple(np.round(theta, 6))

    if key in cache:
        chi2 = cache[key][0]
    else:
        res = compute_chi2(*theta)
        cache[key] = res
        chi2 = res[0]

    if not np.isfinite(chi2):
        return -np.inf

    return lp - 0.5 * chi2

# ============================================================
# Run MCMC
# ============================================================

ndim = 4
nwalkers = 12
nsteps = 50

start = np.array([
    0.24,   # s0
    0.03,   # kc
    0.55,   # n_theta
    3.044,  # ln10As
])

scale = np.array([0.025, 0.010, 0.12, 0.05])
pos = start + scale * np.random.randn(nwalkers, ndim)

# force inside prior
for i in range(nwalkers):
    pos[i,0] = np.clip(pos[i,0], 0.01, 0.34)
    pos[i,1] = np.clip(pos[i,1], 0.012, 0.11)
    pos[i,2] = np.clip(pos[i,2], 0.22, 1.45)
    pos[i,3] = np.clip(pos[i,3], 2.72, 3.28)

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob)

print("="*80)
print("RUN UTIS EXTENDED MCMC")
print("="*80)

sampler.run_mcmc(pos, nsteps, progress=True)

# ============================================================
# Results
# ============================================================

flat = sampler.get_chain(discard=40, thin=2, flat=True)
logp = sampler.get_log_prob(discard=40, thin=2, flat=True)

best_idx = np.argmax(logp)
best = flat[best_idx]
best_res = compute_chi2(*best)

print("\n" + "="*80)
print("BEST FIT")
print("="*80)

print("s0      =", best[0])
print("kc      =", best[1])
print("n_theta =", best[2])
print("ln10As  =", best[3])
print("A_s     =", As_from_ln10As(best[3]))
print("chi2    =", best_res[0])
print("chi2_rsd=", best_res[1])
print("chi2_hz =", best_res[2])
print("chi2_sn =", best_res[3])
print("chi2_bao=", best_res[4])
print("sigma8  =", best_res[5])

summary = pd.DataFrame({
    "param": ["s0", "kc", "n_theta", "ln10As"],
    "mean": np.mean(flat, axis=0),
    "std": np.std(flat, axis=0),
    "p16": np.percentile(flat, 16, axis=0),
    "p50": np.percentile(flat, 50, axis=0),
    "p84": np.percentile(flat, 84, axis=0),
})

display(summary)

# ============================================================
# Save
# ============================================================

chain_df = pd.DataFrame(flat, columns=["s0", "kc", "n_theta", "ln10As"])
chain_df["A_s"] = As_from_ln10As(chain_df["ln10As"].values)
chain_df["logp"] = logp

chain_df.to_csv("/content/utis_extended_mcmc_chain.csv", index=False)
summary.to_csv("/content/utis_extended_mcmc_summary.csv", index=False)

print("\nSaved:")
print("/content/utis_extended_mcmc_chain.csv")
print("/content/utis_extended_mcmc_summary.csv")

# ============================================================
# Corner
# ============================================================

fig = corner.corner(
    flat,
    labels=[r"$s_0$", r"$k_c$", r"$n_\theta$", r"$\ln(10^{10}A_s)$"],
    truths=best,
    show_titles=True,
)

plt.savefig("/content/utis_extended_mcmc_corner.png", dpi=200, bbox_inches="tight")
plt.show()

print("Saved:")
print("/content/utis_extended_mcmc_corner.png")
```

```python
# ============================================================
# UTIS MCMC + Planck-like A_s prior
# (short validation run before full production MCMC)
# ============================================================

!pip -q install emcee corner

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import emcee
import corner

from classy import Class

# ============================================================
# LCDM baseline
# ============================================================

lcdm = Class()

lcdm.set({
    'output': 'mPk',
    'P_k_max_h/Mpc': 5.0,

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,

    'tau_reio': 0.0544,

    'z_max_pk': 3.0
})

lcdm.compute()

sigma8_lcdm = lcdm.sigma8()

print("="*80)
print("LCDM baseline")
print("="*80)

print("sigma8_LCDM =", sigma8_lcdm)

# ============================================================
# DATA
# ============================================================

RSD = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/RSD_clean.csv")
HZ  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/CC_clean.csv")
BAO = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/BAO_clean.csv")
SN  = pd.read_csv("/content/drive/MyDrive/UTIS_project/paper_datasets/SN_pantheon_clean.csv")

# ============================================================
# chi2
# ============================================================

def compute_chi2(params):

    s0, kc, n_theta, ln10As = params

    # -----------------------------
    # priors
    # -----------------------------

    if not (0.0 < s0 < 0.35):
        return -np.inf

    if not (0.01 < kc < 0.12):
        return -np.inf

    if not (0.2 < n_theta < 1.5):
        return -np.inf

    if not (2.6 < ln10As < 3.2):
        return -np.inf

    # -----------------------------
    # convert As
    # -----------------------------

    A_s = np.exp(ln10As) * 1e-10

    # -----------------------------
    # CLASS
    # -----------------------------

    cosmo = Class()

    cosmo.set({
        'output': 'mPk',
        'P_k_max_h/Mpc': 5.0,

        'h': 0.6736,
        'omega_b': 0.02237,
        'omega_cdm': 0.1200,

        'A_s': A_s,
        'n_s': 0.9649,

        'tau_reio': 0.0544,

        'z_max_pk': 3.0
    })

    try:
        cosmo.compute()

    except:
        return -np.inf

    sigma8 = cosmo.sigma8()

    # ========================================================
    # RSD
    # ========================================================

    chi2_rsd = 0.0

    for _, row in RSD.iterrows():

        z = row['z']

        fs8_obs = row['fs8']
        err = row['err']

        try:

            f = cosmo.scale_independent_growth_factor_f(z)

            suppression = (
                1.0
                - s0
                * np.exp(-(kc / (1 + z)))
                * (1/(1+z))**n_theta
            )

            fs8_th = (
                f
                * sigma8
                * suppression
            )

            chi2_rsd += ((fs8_th - fs8_obs)/err)**2

        except:
            return -np.inf

    # ========================================================
    # HZ
    # ========================================================

    chi2_hz = 0.0

    for _, row in HZ.iterrows():

        z = row['z']
        H_obs = row['H']
        err = row['err']

        try:

            H_th = cosmo.Hubble(z) * 299792.458

            chi2_hz += ((H_th - H_obs)/err)**2

        except:
            return -np.inf

    # ========================================================
    # BAO
    # ========================================================

    chi2_bao = 0.0

    for _, row in BAO.iterrows():

        z = row['z']

        try:

            DM = cosmo.angular_distance(z) * (1+z)
            H  = cosmo.Hubble(z) * 299792.458

            rs = cosmo.rs_drag()

            DM_over_rd = DM / rs
            DH_over_rd = 1.0 / (H * rs)

            if not np.isnan(row['DM_over_rd']):

                chi2_bao += (
                    (DM_over_rd - row['DM_over_rd'])
                    / row['DM_over_rd_err']
                )**2

            if not np.isnan(row['DH_over_rd']):

                chi2_bao += (
                    (DH_over_rd - row['DH_over_rd'])
                    / row['DH_over_rd_err']
                )**2

        except:
            return -np.inf

    # ========================================================
    # SN
    # ========================================================

    z = SN['z'].values
    mu_obs = SN['mu'].values
    mu_err = SN['mu_err'].values

    mu_th = []

    try:

        for zz in z:

            dL = cosmo.luminosity_distance(zz)

            mu = 5*np.log10(dL) + 25

            mu_th.append(mu)

    except:
        return -np.inf

    mu_th = np.array(mu_th)

    M_offset = np.mean(mu_obs - mu_th)

    chi2_sn = np.sum(
        ((mu_obs - (mu_th + M_offset))/mu_err)**2
    )

    # ========================================================
    # Planck-like As prior
    # ========================================================

    chi2_As_prior = (
        (ln10As - 3.044)/0.014
    )**2

    # ========================================================
    # total
    # ========================================================

    chi2_total = (
        chi2_rsd
        + chi2_hz
        + chi2_bao
        + chi2_sn
        + chi2_As_prior
    )

    return -0.5 * chi2_total

# ============================================================
# initial walkers
# ============================================================

ndim = 4
nwalkers = 24

p0 = []

for _ in range(nwalkers):

    p0.append([
        np.random.uniform(0.10, 0.30),
        np.random.uniform(0.02, 0.08),
        np.random.uniform(0.30, 1.00),
        np.random.uniform(3.00, 3.08)
    ])

p0 = np.array(p0)

# ============================================================
# run
# ============================================================

print("="*80)
print("RUN UTIS MCMC + Planck prior")
print("="*80)

sampler = emcee.EnsembleSampler(
    nwalkers,
    ndim,
    compute_chi2
)

sampler.run_mcmc(
    p0,
    60,
    progress=True
)

# ============================================================
# chain
# ============================================================

chain = sampler.get_chain(
    discard=20,
    flat=True
)

logp = sampler.get_log_prob(
    discard=20,
    flat=True
)

df = pd.DataFrame({
    's0': chain[:,0],
    'kc': chain[:,1],
    'n_theta': chain[:,2],
    'ln10As': chain[:,3],
    'logp': logp
})

# ============================================================
# summary
# ============================================================

print("="*80)
print("POSTERIOR SUMMARY")
print("="*80)

summary = []

for c in ['s0','kc','n_theta','ln10As']:

    vals = df[c]

    summary.append({
        'param': c,
        'mean': vals.mean(),
        'std': vals.std(),
        'p16': np.percentile(vals,16),
        'p50': np.percentile(vals,50),
        'p84': np.percentile(vals,84),
    })

summary = pd.DataFrame(summary)

display(summary)

# ============================================================
# corner
# ============================================================

fig = corner.corner(
    chain,
    labels=[
        r"$s_0$",
        r"$k_c$",
        r"$n_\theta$",
        r"$\ln(10^{10}A_s)$"
    ],
    show_titles=True
)

plt.show()

# ============================================================
# save
# ============================================================

df.to_csv(
    "/content/utis_planck_prior_chain.csv",
    index=False
)

summary.to_csv(
    "/content/utis_planck_prior_summary.csv",
    index=False
)

print()
print("Saved:")
print("/content/utis_planck_prior_chain.csv")
print("/content/utis_planck_prior_summary.csv")
```

```python
# ============================================================
# UTIS Theta -> mu(k,a) reconstruction
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# BEST-FIT PARAMETERS
# ============================================================

s0 = 0.264
kc = 0.051
n_theta = 0.40

# ============================================================
# GRID
# ============================================================

k_vals = np.logspace(-3, 0, 200)
a_vals = np.linspace(0.2, 1.0, 160)

K, A = np.meshgrid(k_vals, a_vals)

# ============================================================
# THETA(k,a)
# ============================================================

Theta = s0 * (A**n_theta) * (K**2 / (K**2 + kc**2))

# ============================================================
# mu(Theta)
# ============================================================

MU = np.cos(Theta)**2

# ============================================================
# residual suppression
# ============================================================

SUP = 1.0 - MU

# ============================================================
# SUMMARY
# ============================================================

print("="*80)
print("UTIS Theta-sector reconstruction")
print("="*80)

print("s0      =", s0)
print("kc      =", kc)
print("n_theta =", n_theta)

print("\n")
print("Theta min =", Theta.min())
print("Theta max =", Theta.max())

print("mu min =", MU.min())
print("mu max =", MU.max())

# ============================================================
# PLOT : Theta
# ============================================================

plt.figure(figsize=(8,6))

pcm = plt.pcolormesh(
    K,
    A,
    Theta,
    shading='auto'
)

plt.xscale('log')

plt.xlabel("k [h/Mpc]")
plt.ylabel("a")

plt.title("UTIS Theta(k,a)")

plt.colorbar(pcm, label='Theta')

plt.show()

# ============================================================
# PLOT : mu
# ============================================================

plt.figure(figsize=(8,6))

pcm = plt.pcolormesh(
    K,
    A,
    MU,
    shading='auto'
)

plt.xscale('log')

plt.xlabel("k [h/Mpc]")
plt.ylabel("a")

plt.title("UTIS mu(k,a)=cos^2 Theta")

plt.colorbar(pcm, label='mu')

plt.show()

# ============================================================
# PLOT : suppression
# ============================================================

plt.figure(figsize=(8,6))

pcm = plt.pcolormesh(
    K,
    A,
    SUP,
    shading='auto'
)

plt.xscale('log')

plt.xlabel("k [h/Mpc]")
plt.ylabel("a")

plt.title("UTIS suppression = 1 - mu")

plt.colorbar(pcm, label='suppression')

plt.show()

# ============================================================
# EXPORT TABLE
# ============================================================

tbl = pd.DataFrame({
    'k': K.flatten(),
    'a': A.flatten(),
    'Theta': Theta.flatten(),
    'mu': MU.flatten(),
    'suppression': SUP.flatten(),
})

path = "/content/utis_theta_mu_surface.csv"

tbl.to_csv(path, index=False)

print("\nSaved:")
print(path)
```

```python
# ============================================================
# UTIS growth index gamma(z)
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from classy import Class

# ============================================================
# LCDM
# ============================================================

lcdm = Class()

lcdm.set({
    'output': 'tCl,pCl,lCl,mPk',
    'lensing': 'yes',

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,

    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,
})

lcdm.compute()

# ============================================================
# UTIS
# ============================================================

utis = Class()

utis.set({
    'output': 'tCl,pCl,lCl,mPk',
    'lensing': 'yes',

    'h': 0.6736,
    'omega_b': 0.02237,
    'omega_cdm': 0.1200,

    'A_s': 2.1e-9,
    'n_s': 0.9649,
    'tau_reio': 0.0544,

    'P_k_max_h/Mpc': 10.0,
    'z_max_pk': 5.0,

    # ========================================================
    # UTIS BEST FIT
    # ========================================================

    'utis_on': 'yes',

    'utis_s0': 0.264,
    'utis_sigma0': 0.02,
    'utis_kc': 0.051,
    'utis_n_theta': 0.40,
})

utis.compute()

# ============================================================
# z grid
# ============================================================

z_arr = np.linspace(0.0, 2.0, 120)

# ============================================================
# compute gamma
# ============================================================

rows = []

for z in z_arr:

    # --------------------------------------------------------
    # LCDM
    # --------------------------------------------------------

    f_l = lcdm.scale_independent_growth_factor_f(z)

    Om_l = (
        lcdm.Omega0_m() * (1+z)**3
        /
        (lcdm.Hubble(z)/lcdm.Hubble(0))**2
    )

    gamma_l = np.log(f_l) / np.log(Om_l)

    # --------------------------------------------------------
    # UTIS
    # --------------------------------------------------------

    f_u = utis.scale_independent_growth_factor_f(z)

    Om_u = (
        utis.Omega0_m() * (1+z)**3
        /
        (utis.Hubble(z)/utis.Hubble(0))**2
    )

    gamma_u = np.log(f_u) / np.log(Om_u)

    rows.append({
        'z': z,

        'f_LCDM': f_l,
        'f_UTIS': f_u,

        'gamma_LCDM': gamma_l,
        'gamma_UTIS': gamma_u,

        'delta_gamma': gamma_u - gamma_l,
    })

tbl = pd.DataFrame(rows)

# ============================================================
# summary
# ============================================================

print("="*80)
print("UTIS growth-index analysis")
print("="*80)

print("\nLCDM gamma mean:")
print(tbl['gamma_LCDM'].mean())

print("\nUTIS gamma mean:")
print(tbl['gamma_UTIS'].mean())

print("\nDelta gamma mean:")
print(tbl['delta_gamma'].mean())

print("\nMax |delta gamma|:")
print(np.max(np.abs(tbl['delta_gamma'])))

# ============================================================
# plot gamma
# ============================================================

plt.figure(figsize=(8,5))

plt.plot(
    tbl['z'],
    tbl['gamma_LCDM'],
    label='LCDM'
)

plt.plot(
    tbl['z'],
    tbl['gamma_UTIS'],
    label='UTIS'
)

plt.xlabel("z")
plt.ylabel("gamma(z)")

plt.title("Growth index")

plt.legend()

plt.show()

# ============================================================
# delta gamma
# ============================================================

plt.figure(figsize=(8,5))

plt.plot(
    tbl['z'],
    tbl['delta_gamma']
)

plt.axhline(
    0,
    ls='--'
)

plt.xlabel("z")
plt.ylabel("UTIS - LCDM")

plt.title("Delta gamma(z)")

plt.show()

# ============================================================
# growth rate
# ============================================================

plt.figure(figsize=(8,5))

plt.plot(
    tbl['z'],
    tbl['f_LCDM'],
    label='LCDM'
)

plt.plot(
    tbl['z'],
    tbl['f_UTIS'],
    label='UTIS'
)

plt.xlabel("z")
plt.ylabel("f(z)")

plt.title("Growth rate")

plt.legend()

plt.show()

# ============================================================
# save
# ============================================================

path = "/content/utis_gamma_analysis.csv"

tbl.to_csv(path, index=False)

print("\nSaved:")
print(path)
```
